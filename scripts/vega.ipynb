```python

import openpyxl

seen = set()

def parse_xlsx_file(file_path):
    id_data = {}
    workbook = openpyxl.load_workbook(file_path, data_only=True)
    for sheet_name in workbook.sheetnames:
        sheet = workbook[sheet_name]
        last_id = -1
        for i in range(1,sheet.max_row):
            id = str(sheet['A'+str(i)].value)
            if '\xa0' in id:
                id = id.replace('\xa0', '')

            if id != 'None' and id is not None and last_id != id:
                id = int(id)
                last_id = int(id)

            if id != 'None' and id is not None and id in seen: continue

            if id is None or id == 'None':
                id = int(last_id)

            data = sheet['B'+str(i)].value
            if data is not None:
                if id in id_data:
                    id_data[id].append(data.replace('\xa0', ''))
                else:
                    id_data[id] = [data.replace('\xa0', '')]

    return id_data

file_path = '../ВЕГА Саенс.xlsx'
id_data_sci = parse_xlsx_file(file_path)

file_path = '../ВЕГА Про.xlsx'
id_data_pro = parse_xlsx_file(file_path)

file_path = '../ВЕГА ГИС.xlsx'
id_data_gis = parse_xlsx_file(file_path)
```

```python
import pandas as pd
def save_correct_excel(data, filename):
    df = pd.DataFrame.from_dict(data, orient='index', columns=['text', 'author', 'source'])

    # Reset the index to have 'id' as a separate column
    df.reset_index(inplace=True)
    df.rename(columns={'index': 'id'}, inplace=True)

    # Save the DataFrame to an Excel file
    df.to_excel(filename, index=False)

save_correct_excel(id_data_sci, "../ВЕГА Саенс.xlsx")
save_correct_excel(id_data_pro, "../ВЕГА Про.xlsx")
save_correct_excel(id_data_gis, "../ВЕГА ГИС.xlsx")
```

```python
import re

def extract_year(text):
    input_string = text
    pattern = r'\b(201[5-9]|202[0-2])\b'
    # pattern = r'\b(19[5-9]\d|20[0-1]\d|202[0-3])\b'
    years = re.findall(pattern, input_string)
    if not years:
        return -1

    extracted_year = years[0]
    return int(extracted_year)
```

```python
import torch
from transformers import BertTokenizer, BertModel

model = BertModel.from_pretrained("bert-base-multilingual-cased")
tokenizer = BertTokenizer.from_pretrained("bert-base-multilingual-cased")

def get_sentence_embedding(sentence):
    inputs = tokenizer(sentence, return_tensors="pt", truncation=True, padding=True)
    # input_ids = inputs["input_ids"]

    # sentence_token_ids = [i for i, token_id in enumerate(input_ids[0]) if token_id == tokenizer.encode(sentence, add_special_tokens=False)[0]]

    # if not sentence_token_ids:
    #     raise ValueError(f"'{sentence}' not found in '{sentence}'.")

    with torch.no_grad():
        outputs = model(**inputs)
        embeddings = outputs.last_hidden_state.mean(dim=1)

    # sentence_embeddings = [embeddings[0][position].numpy() for position in sentence_token_ids]
    
    return embeddings.numpy()
```

```python
# v2
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
  
def get_similarity(sentence1, sentence2):
    X_list = word_tokenize(sentence1)  
    Y_list = word_tokenize(sentence2) 
    
    sw = stopwords.words('russian')  
    l1 =[];l2 =[] 
    
    X_set = {w for w in X_list if not w in sw}  
    Y_set = {w for w in Y_list if not w in sw} 
        
    rvector = X_set.union(Y_set)  
    for w in rvector: 
        if w in X_set: l1.append(1)
        else: l1.append(0) 
        if w in Y_set: l2.append(1) 
        else: l2.append(0) 
    c = 0
        
    for i in range(len(rvector)): 
            c+= l1[i]*l2[i] 
    cosine = c / float((sum(l1)*sum(l2))**0.5) 
    return cosine

similar_sentences = {}
threshold = 0.5
logs = []
for sentence1 in titles:
    similar_sentences[sentence1] = [sentence1]
    for sentence2 in titles:
        if sentence1 != sentence2:
            # similarity_score = cosine_similarity([embedding1], [embedding2])[0][0]
            similarity_score = get_similarity(sentence1, sentence2)
            logs.append([sentence1, sentence2, similarity_score])
            if similarity_score >= threshold:
                similar_sentences[sentence1].append(sentence2)
                print(similarity_score)


```

```python
# v3
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# titles = ["Title 1", "Title 2", "Title 3", ...]  # Your list of article titles

# Initialize the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the titles to TF-IDF vectors
tfidf_matrix = tfidf_vectorizer.fit_transform(titles)

# Calculate cosine similarity between all pairs of titles
similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)



# Perform K-Means clustering
num_clusters = 3  # You can adjust the number of clusters
kmeans = KMeans(n_clusters=num_clusters)
cluster_labels = kmeans.fit_predict(similarity_matrix)

# Create an empty dictionary to store the clustered titles
clustered_titles = {}
for cluster_id in range(num_clusters):
    clustered_titles[cluster_id] = []

# Populate the dictionary with titles in each cluster
for i, cluster_id in enumerate(cluster_labels):
    clustered_titles[cluster_id].append(titles[i])

# Convert the dictionary to an array of arrays of groups of titles
title_groups = [group for group in clustered_titles.values()]

with open('teest_2_grouped_data_threshold.json', 'w', encoding='utf-8') as grouped_json_file:
    json.dump(title_groups, grouped_json_file, ensure_ascii=False)

# with open('.json', 'w', encoding='utf-8') as grouped_json_file:
#     json.dump(title_groups, grouped_json_file)

# Print the result
# for group in title_groups:
#     print(group)


# Perform hierarchical clustering
# linkage_matrix = linkage(similarity_matrix, method='ward')

# Create a dendrogram to visualize the clusters
# dendrogram(linkage_matrix, labels=titles)
# plt.show()
```
